```python
from ultralytics import YOLO

yolo = YOLO("yolo11n.pt")

import yaml

config  = {
    "path" : "C:\\Users\\User\\OneDrive\\Masaüstü\\YOLO-finetune\\License Detection",
    "train" : "images\\train",
    "val" : "images\\val",
    "names" : {0 : "license_plate"}
}

with open("config.yaml","w") as f:
    yaml.dump(config,f,sort_keys=False)
```

```python
import os

data_path ="C:\\Users\\User\\OneDrive\\Masaüstü\\YOLO-finetune\\License Detection\\labels"

label_paths = []
for folder in os.scandir(data_path):
    if folder.is_dir():
        for file in os.scandir(folder.path):
            if file.name.endswith(".txt"):
                label_paths.append(file.path)
```

```python
classes = [] # all 0's -> license only
for label in label_paths:
    with open(label) as f:
        for line in f:
            curr_class = line.strip().split()[0]
            if curr_class not in classes:
                classes.append(curr_class)
```
